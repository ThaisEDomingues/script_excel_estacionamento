```python
import requests
import json

#para colocar data e hora
from datetime import datetime

#vamos colocar nosso end-point
url = 'https://script.google.com/macros/s/AKfycbw2LAvdIxuWXl8qlX6ylfBStU8uxdnwDH1aSzyi9w8SLXOSp3RZWKOthZBLU8XmaCWP/exec'

#vamos criar a funcao que envia os dados para o excel
def enviar_dados(placa, carro, funcionario):
  #vou colocar aqui data e hora
  data_e_hora = datetime.now().strftime('%d/%m/%y %H:%M:%S')

  #%d -
  #%m - mes
  #y - ano
  #H - hora
  #%M - minuto
  #%S - segundo
  #datetime.now() - salva o momento que o codigo é executado por padrão é americano para tranformar hora portugues
  #usa a funcao dentro do datetime.now().strftime('%d/%m/%y %H:%M:%S')

  parametros = {
      'data_hora': data_e_hora,
      'placa': placa,
      'carro': modelo,
      'funcionario':funcionario
  }
  print('enviando dados para a tabela')

  resposta = requests.post(url, json=parametros)
  if resposta.status_code == 200:
    print('Dados envidados')
  else:
    print(f'erro ao enviar os dados: {resposta.status_code}')




```

```python
while True:
  print('-'*50)
  placa = input('informe a placa do carro: ')
  modelo = input('informe o modelo do carro: ')
  funcionario = input('informe o funcionario que registrou: ')

  enviar_dados(placa, modelo, funcionario)
  op = input('deseja sair? S/N')
  if op == 's':
      break




```
